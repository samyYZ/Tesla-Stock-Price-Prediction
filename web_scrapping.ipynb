```python
# Amazon product review page
base_url = "https://www.flipkart.com/titan-mechanical-automatic-grey-dial-black-leather-strap-analog-watch-men/product-reviews/itm3799ec276b122?pid=WATH47UHUJVWDBY3&lid=LSTWATH47UHUJVWDBY3IWMGWD&marketplace=FLIPKART"
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

import pandas as pd
import time

# 1. Create driver FIRST
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

# 2. Open page
driver.get(base_url)

try:
    close_btn = WebDriverWait(driver, 5).until(
        EC.element_to_be_clickable((By.XPATH, '//button[contains(text(),"✕")]'))
    )
    close_btn.click()
except:
    pass

# 3. THEN wait for elements
WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.XPATH, '//div[contains(@class,"EPCmJX")]'))
)
```

```python
all_reviews = []

for page in range(1, 4):
    url = base_url + f"&page={page}"
    print("Scraping:", url)

    driver.get(url)
    time.sleep(3)

    reviews = driver.find_elements(By.XPATH, '//div[contains(@class,"_1AtVbE")]')

    print("Found reviews:", len(reviews))  # DEBUG

    for r in reviews:
        try:
            name = r.find_element(By.XPATH, './/p[contains(@class,"_2sc7ZR")]').text
            rating = r.find_element(By.XPATH, './/div[contains(@class,"_3LWZlK")]').text
            comment = r.find_element(By.XPATH, './/div[contains(@class,"t-ZTKy")]').text

            all_reviews.append({
                "Customer Name": name,
                "Rating": rating,
                "Comment": comment
            })

        except:
            continue
```

```python
driver.quit()
```

```python
df = pd.DataFrame(all_reviews)
```

```python
df.to_csv("flipkart_reviews.csv", index=False)
print("Saved successfully")
```
